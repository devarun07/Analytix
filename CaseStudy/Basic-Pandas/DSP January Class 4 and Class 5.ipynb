```python
TS_Q = stores.TotalSales.quantile([0,0.01,0.05,0.1,0.25,0.5,0.75,0.9,0.95,0.99,1])
# 0%    1%    5%    10%    25%    50%    75%    90%    95%    99%    100%    
TS_Q
```

```python
stores.TotalSales.min()
```

```python
stores.TotalSales.max()
```

```python
# For symmetric curves
# IQR Rule
# Q3 : 75%
# Q1 : 25%

## IQR = Q3 - Q1

# UC = Q3 + 1.5 * IQR
# LC = Q1 - 1.5 * IQR

Q3 = TS_Q.loc[0.75]
Q1 = TS_Q.loc[0.25]

IQR = Q3 - Q1

UC = Q3 + 1.5 * IQR
LC = Q1 - 1.5 * IQR

print("The upper cap is {} and lower cap is {}".format(UC,LC))

stores.TotalSales.clip_upper(UC)
stores.TotalSales.clip_lower(LC)
```

```python
# Gaussian/Normal Curves

# UC = mean + 3 * SD
# LC = mean - 3 * SD

UC = stores.TotalSales.mean() + 3 * stores.TotalSales.std()
LC = stores.TotalSales.mean() - 3 * stores.TotalSales.std()

print("The upper cap is {} and lower cap is {}".format(UC,LC))

stores.TotalSales.clip_upper(UC)
stores.TotalSales.clip_lower(LC)
```

```python
# Asymetric

# UC = 99%
# LC = 1%

# or,

# UC = 95%
# LC = 5%

UC = TS_Q.loc[0.99]
LC = TS_Q.loc[0.01]

print("The upper cap is {} and lower cap is {}".format(UC,LC))

```
