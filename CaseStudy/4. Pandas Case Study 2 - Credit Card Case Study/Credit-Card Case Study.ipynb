1. In the above dataset

a. In case age is less than 18, replace it with mean of age values.

```python
mean = Customer_Acquisition['Age'].mean()
Customer_Acquisition['Age'][Customer_Acquisition['Age'] < 18] = mean
Customer_Acquisition.head()
```

b. In case spend amount is more than the limit, replace it with 50% of that customer’s limit.
(customer’s limit provided in acquisition table is the per transaction limit on his card)

```python
Merged_Spend = pd.merge(Customer_Acquisition,Spend,on='Customer')
```

```python
def CreditLimit(row):
    if row['Spend_Amount'] > row['Limit']:
        return (row['Limit'] * 0.5)
    else:
        return row['Spend_Amount']
```

```python
Merged_Spend['Spend_Amount'] = Merged_Spend.apply(CreditLimit,axis=1)
Merged_Spend.head()
```

 c. Incase the repayment amount is more than the limit, replace the repayment with the limit. 

```python
Merged_Repayment = pd.merge(Customer_Acquisition,Repayment,on='Customer')
Merged_Repayment.head()
```

```python
def RepaymentLimit(row):
    if row['Repayment_Amount'] > row['Limit']:
        return row['Limit']
    else:
        return row['Repayment_Amount']
```

```python
Merged_Repayment['Repayment_Amount'] = Merged_Repayment.apply(RepaymentLimit, axis=1)
```

```python
Merged_Repayment[Merged_Repayment['Repayment_Amount'] > Merged_Repayment['Limit']].head()
```

e. If the monthly rate of interest is 2.9%, what is the profit for the bank for each month? 

```python
interest_df = pd.merge(Spend,Repayment,on='Customer')
# interest_df[' Monthly Profit '] = interest_df['Repayment_Amount'] - interest_df['Spend_Amount']
```

```python
a = Spend.groupby(by='new_month')['Spend_Amount'].sum()
b = Repayment.groupby(by='new_month')['Repayment_Amount'].sum()
interest_df = pd.concat([a,b],axis=1)
interest_df = interest_df.reset_index()
interest_df['Monthly_Profit'] = interest_df['Repayment_Amount'] - interest_df['Spend_Amount']
interest_df['Profit'] =  interest_df['Monthly_Profit'].apply(lambda x: (x*0.29) if x > 0 else 0)
for month,value in zip(interest_df.new_month,interest_df.Profit):
    print("The profit for the month {} is {}.".format(month,round(value,2)))
```

3. Calculate the city wise spend on each product on yearly basis. Also include a graphical representation for the same.


```python
Merged_Spend['new_year'] = pd.to_datetime(Merged_Spend['Month']).dt.strftime('%y')
df0 = Merged_Spend.groupby(by=['new_year','City','Product'])[['Spend_Amount']].sum().add_prefix('SumOf_')
df0.head(10)
```

```python
for year,df in df0.groupby(level=0):
    df = df.reset_index()
    df = df.pivot(index='City', columns='Product', values='SumOf_Spend_Amount')
    print("Year {}".format(year))
    df.plot(kind='bar')
    plt.title(year)
    plt.ylabel('Spend_Amount')
    plt.show()
```

```python
Merged_Spend['new_month'] = pd.to_datetime(Merged_Spend['Month']).dt.strftime('%m/%y')
x = Merged_Spend.groupby(['new_month','City'])[['Spend_Amount']].sum().add_prefix('SumOf_')
for mon,df in x.groupby(level=0):
    print(mon)
    df = df.loc[mon].reset_index()
    plt.barh(width=df.SumOf_Spend_Amount,y=df.City)
    plt.title(mon)
    plt.show()
```

5. Write user defined PYTHON function to perform the following analysis:

You need to find top 10 customers for each city in terms of their repayment amount by
different products and by different time periods i.e. year or month. The user should be able
to specify the product (Gold/Silver/Platinum) and time period (yearly or monthly) and the
function should automatically take these inputs while identifying the top 10 customers.

```python
Merged_Repayment['new_year'] = pd.to_datetime(Merged_Repayment['Month']).dt.strftime('%Y')
```

```python
def topTenFunction(product, year):
    df1 = Merged_Repayment.groupby(by=['City','Product','new_year','Customer'])[['Repayment_Amount']].sum().sort_values(by='Repayment_Amount',ascending=False)
    for city,df in df1.groupby(level=0):
        print("")
        print("The top ten Cutomers of City: {} with product: {} and year: {}".format(city,product,year))
        for i in df.loc[city,product,year].index:
            print(i)
```
