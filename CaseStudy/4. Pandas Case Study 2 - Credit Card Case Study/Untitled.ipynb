```python
Customer_Acquisition = pd.read_csv('Customer Acqusition.csv')
Customer_Acquisition.head()
```

```python
Spend = pd.read_csv('spend.csv')
Spend = Spend.rename(columns={'Amount': 'Spend_Amount'})
Spend.head()
```

```python
Repayment = pd.read_csv('Repayment.csv')
Repayment = Repayment.rename(columns={'Amount': 'Repayment_Amount'})
Repayment.head()
```

1. In the above dataset

a. In case age is less than 18, replace it with mean of age values.

```python
mean = Customer_Acquisition['Age'].mean()
Customer_Acquisition['Age'][Customer_Acquisition['Age'] < 18] = mean
Customer_Acquisition.head()
```

b. In case spend amount is more than the limit, replace it with 50% of that customer’s limit.
(customer’s limit provided in acquisition table is the per transaction limit on his card)

```python
Merged_Spend = pd.merge(Customer_Acquisition,Spend,on='Customer')
```

```python
def CreditLimit(row):
    if row['Spend_Amount'] > row['Limit']:
        return (row['Limit'] * 0.5)
    else:
        return row['Spend_Amount']
```

```python
Merged_Spend['Spend_Amount'] = Merged_Spend.apply(CreditLimit,axis=1)
Merged_Spend.head()
```

 c. Incase the repayment amount is more than the limit, replace the repayment with the limit. 

```python
Merged_Repayment = pd.merge(Customer_Acquisition,Repayment,on='Customer')
Merged_Repayment.head()
```

```python
def RepaymentLimit(row):
    if row['Repayment_Amount'] > row['Limit']:
        return row['Limit']
    else:
        return row['Repayment_Amount']
```

```python
Merged_Repayment['Repayment_Amount'] = Merged_Repayment.apply(RepaymentLimit, axis=1)
```

```python
Merged_Repayment[Merged_Repayment['Repayment_Amount'] > Merged_Repayment['Limit']].head()
```

 c. What is the average monthly spend by customers? 

```python
Spend['new_month'] = pd.to_datetime(Spend['Month']).dt.strftime('%m/%Y')
```

```python
temp = Spend.groupby(by=['Customer'])['Spend_Amount'].sum() / Spend.groupby(by=['Customer'])['new_month'].count()
for value,index in zip(temp,temp.index):
    print("The average monthly spend of {} is {}.".format(index,round(value,2)))
    
```

 d. What is the average monthly repayment by customers? 

```python
Repayment['new_month'] = pd.to_datetime(Repayment['Month']).dt.strftime('%m/%Y')
```

```python
temp = Repayment.groupby(by=['Customer'])['Repayment_Amount'].sum() / Repayment.groupby(by=['Customer'])['new_month'].count()
for value,index in zip(temp,temp.index):
    print("The average monthly repayment of {} is {}.".format(index,round(value,2)))
```

e. If the monthly rate of interest is 2.9%, what is the profit for the bank for each month? 

```python
interest_df = pd.merge(Spend,Repayment,on='Customer')
# interest_df[' Monthly Profit '] = interest_df['Repayment_Amount'] - interest_df['Spend_Amount']
```

```python
a = Spend.groupby(by='new_month')['Spend_Amount'].sum()
b = Repayment.groupby(by='new_month')['Repayment_Amount'].sum()
interest_df = pd.concat([a,b],axis=1)
interest_df = interest_df.reset_index()
interest_df['Monthly_Profit'] = interest_df['Repayment_Amount'] - interest_df['Spend_Amount']
interest_df['Profit'] =  interest_df['Monthly_Profit'].apply(lambda x: (x*0.29) if x > 0 else 0)
for month,value in zip(interest_df.new_month,interest_df.Profit):
    print("The profit for the month {} is {}.".format(month,round(value,2)))
```
